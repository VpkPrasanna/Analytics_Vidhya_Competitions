```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
```

```python
train = pd.read_csv("train/train.csv")
```

```python
train_image = []
for i in tqdm(range(train.shape[0])):
    img = image.load_img("/home/pcuser/Downloads/janta-hack-CV/train/images/"+train['image_names'][i], target_size=(28,28,1), grayscale=True)
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
X = np.array(train_image)
```

```python
y=train['emergency_or_not'].values
y = to_categorical(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state = 294)
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',kernel_initializer='he_uniform',input_shape=(28,28,1)))
model.add(Conv2D(64, (3, 3), activation='relu',kernel_initializer='he_uniform'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))
model.add(Conv2D(128, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same'))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
model.compile(loss='binary_crossentropy',optimizer="sgd",metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=500,  validation_data=(X_test, y_test))
```

```python
test = pd.read_csv("test_vc2kHdQ.csv")
```

```python
test_image = []
for i in tqdm(range(test.shape[0])):
    img = image.load_img('/home/pcuser/Downloads/janta-hack-CV/train/images/'+test['image_names'][i], target_size=(28,28,1), grayscale=True)
    img = image.img_to_array(img)
    img = img/255
    test_image.append(img)
test = np.array(test_image)
```

```python
prediction = model.predict_classes(test)
```
